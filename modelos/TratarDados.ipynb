```python
def normalizar_imagens(path):
    img = tf.image.decode_image(tf.io.read_file(path))
    img = tf.cast(img, tf.float32)
    img = (img - 128) / 128
    return img
```

```python
def transformar_cinza(path):
    img = tf.image.decode_image(tf.io.read_file(path))
    img = tf.cast(img, tf.float32)
    img = tf.image.rgb_to_grayscale(img)
    return img
```
