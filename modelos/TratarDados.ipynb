```python
import os

def mostrar_arquivos(path):
    for p, _, files in os.walk(os.path.abspath(path)):
        for file in files:
            print(file)
```

```python
import os

dicionario_cancer = {'SOB_B_A': 'A', 'SOB_B_F':'F','SOB_B_PT':'PT','SOB_B_TA':'TA','SOB_M_DC': 'DC', 
                     'SOB_M_LC':'LC','SOB_M_MC':'MC','SOB_M_PC':'PC'}

def mudar_label(path):
    x = 0
    for p, _, files in os.walk(os.path.abspath(path)):
        for file in files:
            x += 1
            os.rename(p+"\\"+file,p+"\\"+dicionario_cancer.get(file.split('-')[0], 'Contato n√£o encontrado')+f'{x}.png')
```

```python
mudar_label(r'C:\Users\rodrigo.cabral\PycharmProjects\TCC-CienciaDados\Imagens\400X_treino')
```

```python
import os, random
imagens = []
dirB = r'C:\Users\rodrigo.cabral\PycharmProjects\TCC-CienciaDados\Imagens\100X_treino\B'
dirM = r'C:\Users\rodrigo.cabral\PycharmProjects\TCC-CienciaDados\Imagens\100X_treino\M'
for x in range(0,2):
    imagens.append(f'{dirB}\\{random.choice(os.listdir(dirB))}')
    imagens.append(f'{dirM}\\{random.choice(os.listdir(dirM))}')
```

```python
import matplotlib.pyplot as plt 
fig, axes = plt.subplots(nrows=2, ncols=2)
cmaps = [0,1,2,3]
for ax, imagem in zip(axes.flat, imagens):
    img = plt.imread(imagem)
    ax.imshow(img)
    ax.set(title=imagem.split('\\')[-1].replace('.png',''))
    ax.axis('off')

plt.show()
```
